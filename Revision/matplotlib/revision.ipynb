```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line chart is for showing trends over time
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]


plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='^', label='Data', linestyle='--', color='blue', linewidth=1)
plt.title('Line Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Bar chart is used for to compare the different categories


Subjects = ["Math", "English", "Hindi", "Science", "Social", "Computer"]
SubjectsMarks = [90, 80, 70, 60, 50, 98]



plt.bar(Subjects, SubjectsMarks, color='skyblue', label='Marks', width=0.5, align='center', edgecolor='black')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Comparing Marks')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


```

```python
# Pie chart is used to show the proportions of different categories


Subjects = ["Math", "English", "Hindi", "Science", "Social", "Computer"]



plt.pie(SubjectsMarks, labels=Subjects, autopct='%0.1f%%', shadow=True, startangle=90)
plt.title('Pie Chart')
plt.show()
```
