```python
import pandas as pd
import numpy as np

from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# Simple dataset
data = {
    "Size": [800, 900, 1000, 1100, 1200],
    "Rooms": [2, 2, 3, 3, 4],
    "Age": [20, 15, 10, 8, 5],
    "Price": [80, 90, 110, 130, 150]
}                    
  

df = pd.DataFrame(data)

print(f"Shape of the Data is : {df.shape}")

X = df.iloc[:,:-1]
y = df["Price"]
  
print(f"X: \n {X}")
print(f"y: \n {y}")                    
                      
                      
                      
                      
                      
  

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score



lr = LinearRegression()

fs = SFS(lr, k_features=2, forward=True, floating=False, scoring='r2', cv=0)
lr.fit(X, y)
```

```python
from sklearn.linear_model import LinearRegression


# Your Linear Regression model
lr = LinearRegression()

# Sequential Feature Selector
fs = SFS(lr, k_features=3, forward=True, floating=False, scoring='r2', cv=0)
fs = fs.fit(X, y)

# Get the selected feature names
print("Selected features:", fs.k_feature_names_)

# want to see the score
print("score :", fs.k_score_)
```
