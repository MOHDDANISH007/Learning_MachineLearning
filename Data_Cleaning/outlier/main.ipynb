```python
ApplicantIncome = df['ApplicantIncome']
ApplicantIncome.head()


sns.boxplot(df['ApplicantIncome'])

plt.show()


# sns.displot(df['ApplicantIncome'],kde=True,bins=15)



print(df['ApplicantIncome'].unique())
print(df['ApplicantIncome'].describe())


q1 = df['ApplicantIncome'].quantile(0.25)
q3 = df['ApplicantIncome'].quantile(0.75)

print(f"Q1: {q1}\nQ3: {q3}")

iqr = q3 - q1

lower_range = q1 - 1.5 * iqr
upper_range = q3 + 1.5 * iqr

print(f"Lower: {lower_range}\nUpper: {upper_range}")

df['ApplicantIncome'] = df['ApplicantIncome'].clip(lower,upper)
```
