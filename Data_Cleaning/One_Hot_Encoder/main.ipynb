```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer

# Text data (list of strings)
data = [
    "People watch dswithbappy",
    "dswithbappy watch dswithbappy",
    "people write comments",
    "comments write people",
    "dswithbappy write comments"
]

# ---------- OneHotEncoder (sentence-level) ----------
encoder = OneHotEncoder(sparse_output=False)

# [[d] for d in data] doing this thing for make 2d array , so actually it look like [[d],[d],[d],[d],[d]]
encoded = encoder.fit_transform([[d] for d in data])

print("Unique sentences (NOT words):")
print(encoder.categories_)

print("\nOne-hot encoded sentences:")
print(encoded)




# BOW = Bag of Words



sentences = [
"I love AI, I love India",
"I love ML"
]


vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(sentences)

# i want to see the key value pairs data

print(f"Vocabulary:\n {vectorizer.vocabulary_}")

print(f"Bag of words:\n {bow.toarray()}")
```
