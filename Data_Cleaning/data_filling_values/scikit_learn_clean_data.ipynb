```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Our messy dataset
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 35],
    'City': ['New York', np.nan, 'London', 'London']
})

print("Original DataFrame:")
print(df)

# --- STEP 1: Handle Numerical Data (Age) ---
num_imputer = SimpleImputer(strategy='mean')
df['Age'] = num_imputer.fit_transform(df[['Age']]).ravel()


# --- STEP 2: Handle Categorical Data (City) ---
cat_imputer = SimpleImputer(strategy='most_frequent')
df['City'] = cat_imputer.fit_transform(df[['City']]).ravel()

print("\nAfter Imputation:")
print(df)
```
