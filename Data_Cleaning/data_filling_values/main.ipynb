```python
# Import numpy for numerical operations
import numpy as np

# Import pandas for data handling and analysis
import pandas as pd

# Import matplotlib for data visualization (graphs)
import matplotlib.pyplot as plt


# Read the CSV file and load it into a DataFrame
df = pd.read_csv(r'C:/Users/mohdd/OneDrive/Documents/Desktop/GenAI/Data_Cleaning/loan_data.csv')

# Show the first 5 rows of the dataset
df.head()

# Check how many missing (null) values are present in each column
df.isnull().sum()

# Get statistical summary of numerical columns (mean, min, max, etc.)
df.describe()

# Get basic information about the dataset
# (data types, non-null count, memory usage)
df.info()

# Display all column names in the dataset
df.columns

# Find the total number of rows in the dataset
len(df)

# Calculate percentage of missing values in each column
df.isnull().sum() / len(df) * 100

# Get the total count of missing values in the entire dataset
print(df.isnull().sum().sum())


# ------------------------------
# Overall missing value percentage in the entire dataset
# ------------------------------

# Total missing values
total_missing = df.isnull().sum().sum()

# Total number of cells (rows  columns)
total_cells = df.shape[0] * df.shape[1]

# Overall percentage of missing values
overall_missing_percentage = (total_missing / total_cells) * 100

print(overall_missing_percentage)


# plotting missing values


# missing and not missing values
# Count missing values per column
missing_value_count = df.isnull().sum()

# Count not-missing values per column
not_missing_value_count = df.notnull().sum()

# Plot stacked bar chart
plt.figure(figsize=(10, 6))

plt.bar(
    missing_value_count.index,
    missing_value_count.values,
    label='Missing Values'
)

plt.bar(
    not_missing_value_count.index,
    not_missing_value_count.values,
    bottom=missing_value_count.values,
    label='Not Missing Values'
)

plt.xlabel('Columns')
plt.ylabel('Count')
plt.title('Missing vs Not Missing Values in Each Column')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# drop missing values


```
