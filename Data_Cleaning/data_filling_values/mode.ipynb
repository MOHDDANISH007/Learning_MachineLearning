```python
import pandas as pd
import numpy as np

# Create a sample dataset with mixed types
data = {
    'City': ['London', 'Paris', 'London', np.nan, 'Tokyo'], # Categorical
    'Color': ['Red', np.nan, 'Blue', 'Red', 'Red'],        # Categorical
    'Price': [100, 150, np.nan, 200, 250],                 # Numerical
    'Gender': ['F', 'M', np.nan, 'F', 'F']                 # Categorical
}

df = pd.DataFrame(data)

# 1. Select only the categorical columns (usually called 'object' in pandas)
cat_cols = df.select_dtypes(include=['object']).columns
print(f"Categorical columns found: {list(cat_cols)}")

# 2. Loop through those columns and fill with Mode
for col in cat_cols:
    # We use [0] because .mode() returns a Series, and we want the first value
    df[col] = df[col].fillna(df[col].mode()[0])

print("\n--- Data after Categorical Mode Filling ---")
print(df)


df.isnull().sum()
```
