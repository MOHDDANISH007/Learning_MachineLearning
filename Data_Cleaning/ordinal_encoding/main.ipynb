```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

df = pd.DataFrame({
    'color': ['red', 'green', 'blue', 'red', 'green', 'blue',
              'cyan', 'magenta', 'yellow', 'black']
})

# Check unique values
print(df['color'].unique())

# Correct ordinal order (with commas)
ordinal_order = [['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'black']]

# Create encoder
ordinal_encoder = OrdinalEncoder(categories=ordinal_order)

# Fit and transform
df['color_encoded'] = ordinal_encoder.fit_transform(df[['color']])

print(df)
```
