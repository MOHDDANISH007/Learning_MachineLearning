```python
import numpy as np
import pandas as pd
import re
import string

df = pd.read_csv(
    r'C:/Users/mohdd/OneDrive/Documents/Desktop/GenAI/data_preprocessing/IMDB Dataset.csv'
)

print(f"shape of df is {df.shape}")
print(f"columns of df are {df.columns}")

# take only 100 rows
df = df.head(100)

# lowercase
df['review'] = df['review'].str.lower()

def remove_html_tags(text):
    clean = re.compile(r'<[^>]+>')
    return re.sub(clean, '', text)

def remove_url_text(text):
    clean = re.compile(r'https?://\S+|www\.\S+')
    return re.sub(clean, '', text)

def remove_punctuations(text):
    # str.maketrans('a-z', 'A-Z') means replace a-z with A-Z
    # str.maketrans('', '', string.punctuation) means remove all punctuation
    # text.translate(str.maketrans('a-z', 'A-Z')) means replace a-z with A-Z
    # text.translate(str.maketrans('', '', string.punctuation)) means remove all punctuation
    return text.translate(str.maketrans('', '', string.punctuation))

# âœ… correct order
df['review'] = df['review'].apply(remove_html_tags)
df['review'] = df['review'].apply(remove_url_text)
df['review'] = df['review'].apply(remove_punctuations)

print(df.loc[3, 'review'])
```

```python
chatShortForms = {
    "hi": "hello",
    "hru": "how are you",
    "wbu": "what about you",
    "sup": "what's up",
    "pls": "please",
    "thx": "thanks",
    "ty": "thank you",
    "u": "you",
    "ur": "your / you are",
    "r": "are",
    "b4": "before",
    "gr8": "great",
    "btw": "by the way",
    "idk": "i don't know",
    "ik": "i know",
    "omg": "oh my god",
    "lol": "laughing out loud",
    "brb": "be right back",
    "afk": "away from keyboard",
    "ttyl": "talk to you later",
    "gn": "good night",
    "gm": "good morning",
    "tc": "take care",
    "np": "no problem",
    "asap": "as soon as possible"
}

def replace_short_forms(text):
    new_text = []
    for word in text.split():
        if word in chatShortForms:
            new_text.append(chatShortForms[word])
        else:
            new_text.append(word)
    return " ".join(new_text)





# handle incorrect text
# def handle_incorrect_text(text):
#     from textblob import TextBlob
#     textblob = TextBlob(text)
#     return textblob.correct().string


# user_input = "hi give me this report ASAP"

# expanded_text = replace_short_forms(user_input.lower())
# print(expanded_text)

# corrected_text = handle_incorrect_text(expanded_text)
# print(corrected_text)
```
