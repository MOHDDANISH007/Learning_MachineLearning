```python
from nltk.corpus import stopwords
import nltk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(
    r'C:/Users/mohdd/OneDrive/Documents/Desktop/GenAI/data_preprocessing/IMDB Dataset.csv')

df = df.head(100)

# columns


df.columns

# shape
# df.shape


print(f'review the 3rd review: {df["review"][3]}')

# i want to lower case all the data
df['review'] = df['review'].str.lower()
# print(f'review the 3rd review: {df["review"][3]}')


def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    # match all the html tags
    clean = re.compile('<.*?>')
    # replace the html tags with " " in the text
    return re.sub(clean, '', text)


def remove_url(text):
    """Remove url from a string"""
    import re
    # match all the url
    clean = re.compile(
        'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    # replace the url with " " in the text
    return re.sub(clean, '', text)


# remove the punctuations
def remove_punctuations(text):
    """Remove punctuations from a string"""
    import string
    # remove all the punctuations
    return text.translate(str.maketrans('', '', string.punctuation))


df['review'] = df['review'].apply(remove_punctuations)
print(f'review the 3rd review: {df["review"][3]}')


# means it will remove all the html tags
df['review'] = df['review'].apply(remove_html_tags)
print(f'review the 3rd review: {df["review"][3]}')
# print(f'review the 3rd review: {df["review"]}')


# removing  the url from the text
df['review'] = df['review'].apply(remove_url)
print(f'review the 3rd review: {df["review"][3]}')


# # remove the html tags
# df['review'] = df['review'].str.replace('<br />', ' ')

# print(f"review the 3rd review: {df['review'][3]}")


# remove the stop words

# download the stopwords
nltk.download('stopwords')

# print(stopwords.words('english'))

# removing the stop words


def remove_stopwords(text):
    """Remove the stopwords from a string"""
    new_text = []
    split_text = text.split()
    # print(split_text)

    for word in split_text:
        word = word.lower()
        if word not in stopwords.words('english'):
            new_text.append(word)
    return " ".join(new_text)


df['review'] = df['review'].apply(remove_stopwords)
print(f'review the 3rd review: {df["review"][3]}')
```
