```python
import matplotlib.pyplot as plt

x_Days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
y_sales_of_cake = [20, 30, 40, 20, 30, 70, 80]

plt.plot(x_Days, y_sales_of_cake, color="blue", linestyle="--", marker="o", linewidth=1, label="Cake Sales")

plt.title("Sales of Cake")
plt.xlabel("Days")
plt.ylabel("Sales Per Day")
plt.legend()   # âœ… Now it will show 'Cake Sales' in the legend
plt.grid(True)
plt.xlim("Monday", "sunday")
plt.ylim(0,100)
plt.show()           

```

```python
# What is Subplot in matplotlib?


import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]


# you just define first how many rows and columns you want to divide your figure in

# Now when you define the figure then tell on which position you want the graph
plt.subplot(1, 3, 1)
plt.plot(x, y, marker='o', color='blue', linewidth=2)
plt.title('Line Plot')

plt.subplot(1, 3, 2)
plt.bar(x, y, color='green')
plt.tight_layout()
plt.title('Bar chart')


plt.subplot(1, 3, 3)
plt.bar(x, y, color='Red')
plt.tight_layout()
plt.title('Another Bar chart')




```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]


# you just define first how many rows and columns you want to divide your figure in

# Now when you define the figure then tell on which position you want the graph
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

fig.tight_layout()
fig.suptitle('Subplots Example', fontsize=16)

ax[0].plot(x, y, marker='o', color='blue', linewidth=2)
ax[0].set_title('Line Plot')

ax[1].bar(x, y, color='green')
ax[1].set_title('Bar chart')

ax[2].bar(x, y, color='Red')
ax[2].set_title('Another Bar chart')

```
