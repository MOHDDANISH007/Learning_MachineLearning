```python
import numpy as np

# doing manually featurescalling (Normalization)

# Age and Salary dataset
data = np.array([
    [22, 20000],
    [25, 25000],
    [30, 40000],
    [35, 60000],
    [40, 100000]
])

print("Original Data:\n", data)


min_vals = data.min(axis=0)
max_vals = data.max(axis=0)

print("Min Values:\n", min_vals)
print("Max Values:\n", max_vals)

normalized_data = (data - min_vals) / (max_vals - min_vals)

print("Normalized Data:\n", normalized_data)
```

```python
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt

# doing manually featurescalling (Normalization)

# Age and Salary dataset
data = np.array([
    [22, 20000],
    [25, 25000],
    [30, 40000],
    [35, 60000],
    [40, 100000]
])

print("Original Data:\n", data)

# plot the graph before and after normalization

plt.scatter(data[:, 0], data[:, 1])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()

# min_max_scaler = preprocessing.MinMaxScaler()
normalization = preprocessing.MinMaxScaler()

normalized_data = normalization.fit_transform(data)

print("Normalized Data:\n", normalized_data)

# plot the graph after normalization

plt.scatter(normalized_data[:, 0], normalized_data[:, 1])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
