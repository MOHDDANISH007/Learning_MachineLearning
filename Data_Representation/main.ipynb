```python
import re
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB

from sklearn.naive_bayes import MultinomialNB
# import randomforest
from sklearn.ensemble import RandomForestClassifier




# download stopwords
nltk.download('stopwords')

# load data
df = pd.read_csv(
    r'C:/Users/mohdd/OneDrive/Documents/Desktop/GenAI/Data_Representation/IMDB Dataset.csv'
)

df = df.iloc[:10000]
df.drop_duplicates(inplace=True)

# remove HTML tags
def remove_html_tags(text):
    return re.sub('<.*?>', '', text)

df['review'] = df['review'].apply(remove_html_tags)
df['review'] = df['review'].str.lower()

# remove stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join(word for word in text.split() if word not in stop_words)

df['review'] = df['review'].apply(remove_stopwords)

# ✅ FIXED HERE
x = df['review']          # ✔ pandas Series (NOT .values)
y = df['sentiment']

# label encoding
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# train-test split
x_train, x_test, y_train, y_test = train_test_split(
    x, y_encoded, test_size=0.2, random_state=42
)

# Bag of Words
cv = CountVectorizer()
x_train_bow = cv.fit_transform(x_train).toarray()
x_test_bow = cv.transform(x_test).toarray()

# # print the array
# print(x_train_bow)
# print(x_test_bow)

print(x_train_bow.shape)
print(x_test_bow.shape)

# train model
model = GaussianNB()
model.fit(x_train_bow, y_train)

# predict
y_pred = model.predict(x_test_bow)

# accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')



# Train Multinomial Naive Bayes
# this is a classification algorithm
model = MultinomialNB()
model.fit(x_train_bow, y_train)

# Predict on test data
y_pred = model.predict(x_test_bow)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_accuracy:.4f}")

# Predict on training data
y_train_pred = model.predict(x_train_bow)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.4f}")


# ---------------------------Random Forest--------------------------------
randomMode = RandomForestClassifier()
randomMode.fit(x_train_bow, y_train)

# Predict on test data
y_pred = randomMode.predict(x_test_bow)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Algorithm Test Accuracy: {test_accuracy:.4f}")

# Predict on training data
y_train_pred = randomMode.predict(x_train_bow)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Random Forest Algorithm Training Accuracy: {train_accuracy:.4f}")
```
